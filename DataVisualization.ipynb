```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from itertools import cycle, islice
from decimal import Decimal

co2sector = pd.read_csv("D:/Anitha/Graduate/Courses/GradProject/CPSC597/Data/VisualizationData/CleanCO2_Sector.csv")
totalco2 = pd.read_csv("D:/Anitha/Graduate/Courses/GradProject/CPSC597/Data/VisualizationData/2019CO2_state.csv")
sectorco2 = pd.read_csv("D:/Anitha/Graduate/Courses/GradProject/CPSC597/Data/VisualizationData/2019CO2_sector.csv")
top5co2 = pd.read_csv("D:/Anitha/Graduate/Courses/GradProject/CPSC597/Data/VisualizationData/Top5CO2.csv")

co2sector.head()
```

```python
df = pd.DataFrame(totalco2) 
State = df['State']
CO2 = df['Total'] 
# Figure Size
fig = plt.figure(figsize =(8, 5)) 
# Horizontal Bar Plot
plt.xticks(rotation=90)
plt.title("2019 CO2 emissions by State")
plt.bar(State, CO2)
 # Show Plot
plt.show()
```

```python
df = pd.DataFrame(sectorco2) 
Sector = df['Sector']
CO2 = df['Totalemissions'] 
# Figure Size
fig = plt.figure(figsize =(8, 5)) 
plt.title("2019 CO2 emissions by Sector")
plt.bar(Sector, CO2)
 # Show Plot
plt.show()
```

```python
my_colors = list(islice(cycle(['black', 'red', 'blue', 'green','yellow']), None, len(co2sector))) 
bar_graph = co2sector.plot(kind='bar', stacked=True, color=my_colors, figsize=(10, 6))

bar_graph.set_title("State vs Sectors")
bar_graph.set_xlabel("State") 
bar_graph.set_ylabel("CO2 Emission (MMT)") 
plt.grid(b=True, which='major', color='grey', linestyle=':')

for x in bar_graph.patches: 
    if x.get_height() >= 50: 
        bar_graph.annotate(str(int(x.get_height())), (x.get_x() * 1, x.get_height() * 1))

plt.xticks(rotation=70) 
#plt.xticklabels((rotation=70)) 
plt.tight_layout()
```
